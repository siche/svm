```python
# -*-coding:utf-8 -*-
from mysvm import *

# Load data  

train_datas=load_image('Data/train_im.idx3-ubyte')
train_labels=load_label('Data/train_label.idx1-ubyte')

test_datas=load_image('Data/test_im.idx3-ubyte')
test_labels=load_label('Data/test_label.idx1-ubyte')

train_num=np.size(train_datas,0)
test_num=np.size(test_datas,0)

train_hist=np.zeros((train_num,128),dtype=np.float32)
test_hist=np.zeros((test_num,128),dtype=np.float32)

for i in range(train_num):
       c=hog(deskw(train_datas[i][:]))
       train_hist[i,:]=c

for k in range(test_num):
       test_hist[k][:]=hog(deskw(test_datas[k][:]))

def svm_test(choice=1):
	# choice =1 Linear Model
	# choice =2 RBF model
	# choice =3 Hog + Linear


	#  Grid Search
	if choice==1:
		c=np.linspace(1,20,2)
		gamma=np.linspace(1,10,2)
		best_c,best_g=grid_search(c, gamma, train_datas[:10000][:], train_labels[:10000], test_datas, test_labels,1)
	if choice==2:
		gamma=np.linspace(0.01,0.050,2)
		c = np.linspace(15,45,2)
		best_c,best_g=grid_search(c, gamma, train_datas[:10000][:], train_labels[:10000], test_datas, test_labels,1)


	# Train and Test SVM with different size of train data
	train_num=np.linspace(10000,60000,6,dtype=int)
	acc=np.zeros(6)
	result=np.zeros((6,10000),dtype=int)
	cost=np.zeros(6)

	for i in range(6):
		num=train_num[i]
		if choice < 3:
			train_data=train_datas[:num][:]
			test_data=test_datas
			model=svm(best_c,best_g,choice)

		if choice==3:
			model=svm(c=2.67,gamma=5.382,kernel=1)
			train_data=train_hist[:num][:]
			test_data=test_hist

		train_label=train_labels[:num]
		test_label=test_labels

		# Get Data
		t=model.train(train_data,train_label)
		res, acca=model.predict(test_data,test_label)

		acc[i]=acca
		cost[i]=t
		result[i][:]=np.reshape(res,-1)

		model.save('choice=%r_%r.xml' % (choice,num))
		print ('%r/6 Finished \n Accuracy of choice %r=%r' %( (i+1), choice, 100*acca))
	return acc, cost , result

acc_1,cost1, res1=svm_test(1)
acc_2,cost2, res2=svm_test(2)
acc_3,cost3, res3=svm_test(3)
```
